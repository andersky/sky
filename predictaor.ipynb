```python
import gradio as gr  
import pandas as pd  
import pickle  
  
# 加载模型  
with open('D:/HuaweiMoveData/Users/zero/Desktop/Titantic.pkl', 'rb') as f:  
    model = pickle.load(f)  
  
# 定义预测函数（稍作修改以适应新的输入格式）  
def predict_survival(PassengerId, sex, age, sibsp, parch, fare, embarked, pclass, family):  
    # 创建单个样本的 DataFrame  
    df = pd.DataFrame({  
        'PassengerId': [PassengerId],  # 假设 PassengerId 是整数，不需要转换  
        'Sex': [1 if sex == 'male' else 0],  # 直接使用字符串  
        'Age': [float(age)],  # 假设 age 是字符串表示的浮点数  
        'SibSp': [sibsp],  # 假设 SibSp 是整数  
        'Parch': [parch],  # 假设 Parch 是整数  
        'Fare': [float(fare)],  # 假设 Fare 是字符串表示的浮点数  
        # 假设模型是用独热编码处理 Embarked 和 Pclass 的  
        'Embarked_C': [1 if embarked == 'C' else 0],  
        'Embarked_Q': [1 if embarked == 'Q' else 0],  
        'Embarked_S': [1 if embarked == 'S' else 0],  
        'Pclass_1': [1 if pclass == '1' else 0],  
        'Pclass_2': [1 if pclass == '2' else 0],  
        'Pclass_3': [1 if pclass == '3' else 0],  
        'family': [family]  # 假设 FamilySize 是整数或浮点数，根据具体情况转换  
    })  
    # 使用模型获取存活的概率  
    probabilities = model.predict_proba(df)  
    survival_probability = probabilities[0][1]  # 获取存活类别的概率  
    return f"Survival Probability: {survival_probability:.2%}"  
  
# 创建 Gradio 界面  
iface = gr.Interface(  
    fn=predict_survival,  
    inputs=[  
        gr.Number(label="PassengerId"),
        gr.Dropdown(choices=['male', 'female'], label="Sex"),  
        gr.Number(label="Age"),  
        gr.Number(label="Number of Siblings/Spouses"),  
        gr.Number(label="Number of Parents/Children"),  
        gr.Number(label="Fare"),  
        gr.Radio(choices=['C', 'Q', 'S'], label='Embarked'),  
        gr.Radio(choices=['1', '2', '3'], label='Pclass'),  
        gr.Number(label="family")  
    ],  
    outputs=gr.Text(label="Prediction"),  
    title="Titanic Survival Prediction",  
    description="Input the details to predict the survival probability on the Titanic."  
)  

# 启动应用  
iface.launch(share=True) 
```
